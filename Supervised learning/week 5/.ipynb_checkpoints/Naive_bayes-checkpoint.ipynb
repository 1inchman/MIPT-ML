```python
from sklearn.datasets import load_digits, load_breast_cancer
from sklearn.cross_validation import cross_val_score
from sklearn import naive_bayes as nb
```

```python
X_digits = load_digits().data
y_digits = load_digits().target

X_cancer = load_breast_cancer().data
y_cancer = load_breast_cancer().target

# print(X_digits[:10, :])
print(X_cancer[:10, :])
```

```python
bernoulli = nb.BernoulliNB()
score = cross_val_score(estimator = bernoulli, X=X_digits, y=y_digits, cv=3)
print(score.mean())
multi = nb.MultinomialNB()
score = cross_val_score(estimator = multi, X=X_digits, y=y_digits, cv=3)
print(score.mean())
gauss = nb.GaussianNB()
score = cross_val_score(estimator = gauss, X=X_digits, y=y_digits, cv=3)
print(score.mean())
```

```python
bernoulli = nb.BernoulliNB()
score = cross_val_score(estimator = bernoulli, X=X_cancer, y=y_cancer, cv=3)
print(score.mean())
multi = nb.MultinomialNB()
score = cross_val_score(estimator = multi, X=X_cancer, y=y_cancer, cv=3)
print(score.mean())
gauss = nb.GaussianNB()
score = cross_val_score(estimator = gauss, X=X_cancer, y=y_cancer, cv=3)
print(score.mean())
```
